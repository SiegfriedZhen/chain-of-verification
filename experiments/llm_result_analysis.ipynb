```python
import ast
import json

import pandas as pd
```

```python
file = '../data/knowledge_base_v2.xlsx'
cove_file = '../results/cove_results_20250419_234623.xlsx'
raw_data_file = '../data/yt_tsai_secret.xlsx'

df = pd.read_excel(file)
content_df = pd.read_excel(cove_file)
raw_data_df = pd.read_excel(raw_data_file)

task_name = 'youtube_search_蔡英文秘史_1391000e'
df = df[df['task_name'] == task_name]
```

```python
import re, ast, json

raw = content_df[content_df.iteration == 20].final_assessment.values[0]

# 1. 用正規表達式抓 content='…' 裡面的內容
m = re.search(r"content='(?P<body>.*)' additional_kwargs=", raw)
if not m:
    raise ValueError("找不到 content 區段")

body_literal = m.group('body')

# 2. 用 ast.literal_eval 讓 Python 幫你反解 escape sequences
#    它會把 \n, \", \uXXXX 都處理好，而且原本就是真正的中文不會被動
content_str = ast.literal_eval(f"'{body_literal}'")

# 3. 交給 json.loads
data = json.loads(content_str)
for k, v in data.items():
    print(k)
    print(v, '\n')
```

```python

detail_df = pd.DataFrame(
[ ast.literal_eval(x) for x in df[df['content_type'] == 'evaluation_result']['content']]
)

for round, row in enumerate(detail_df.itertuples()):
    print(f"Iteration {round+1}\n: {row.execution_result}")
    print(row.target_technique)
    print(row.target_evidence)
    print(f"found_evidence:\n{row.found_evidence}")
    print(f"reasoning:\n{row.reasoning}")
    print(f"final_reward_score: {row.final_reward_score}")
    print('-'*100)
```
